```python
from ultralytics import YOLO

# Choosing pre-trained model
model = YOLO('yolov8n.pt')

results = model.train(data='data.yml', epochs=30, imgsz=640, batch=16)
```

```python
import matplotlib.pyplot as plt
import cv2

# Loading the best checkpoint
model = YOLO("runs/detect/train/weights/best.pt")

#Run on a single image
image_path = "test_image/613.jpg" ## Create this folder if not exist & place the image
frame = cv2.imread(image_path)

results = model(image_path)
result = results[0]

annotated_image = result.plot()
annotated_image_rgb = cv2.cvtColor(annotated_image, cv2.COLOR_BGR2RGB)

plt.imshow(annotated_image_rgb)
plt.axis('off')
plt.show()
```

```python
if result:
  # Extracting the first bounding box
  box = result.boxes.xyxy[0].cpu().numpy().astype(int)
  x1, y1, x2, y2 =  box

  # Cropping detected license plate
  plate_image = frame[y1:y2, x1:x2]

  # Display the cropped image
  plt.imshow(plate_image)
  plt.axis("off")
  plt.show()
else:
  print("No License Plate detected.")
```

```python
import easyocr

# Reading character from License plate
reader = easyocr.Reader(['en'], gpu=False)
ocr_result = reader.readtext(plate_image)
text = " ".join([res[1] for res in ocr_result])
print("Detected Text:", text)
```

```python
# Alternate to EasyOCR
import pytesseract

gray = cv2.cvtColor(plate_image, cv2.COLOR_BGR2GRAY)
_, thresh = cv2.threshold(gray, 150, 255, cv2.THRESH_BINARY)
text = pytesseract.image_to_string(thresh)
print("Detected Plate:", text)
```
